#### Within‚Äëgroup agreement

```python
import pandas as pd
import numpy as np
from math import log2

k = 5  
expected_var_uniform = (k ** 2 - 1) / 12  


def consensus_stats(series):
    vals = series.dropna().to_numpy()
    n = len(vals)
    mean = vals.mean()
    sd = vals.std(ddof=1)
    var = sd ** 2

    # r_wg (within‚Äëgroup agreement; James, Demaree & Wolf 1984) => higher score: higher agreement
    rwg = 1 - (var / expected_var_uniform)
    rwg = max(rwg, 0)  


    return pd.Series(
        {
            "Mean": mean,
            "SD": round(sd, 3),
            "r_wg": round(rwg, 3),
        }
    )


results = responses.apply(consensus_stats, axis=0).T
results

#==> moderate (simplification) to high agreement (others) based on r_wg index scores
```
